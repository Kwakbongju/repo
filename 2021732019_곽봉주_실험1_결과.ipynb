# 2. 실험에 관한 고찰  
Markdown을 사용하며 기존의 문서와 사용법의 차이에서 약간의 불편함을 느꼈다.. 줄바꿈이 enter를 누른다고 되는 것이 아닌 스페이스바 두 번을 한 뒤에 엔터키를 눌러야 했기 때문이다. 또한 이미지를 직접적으로 사용하지 않는다는 것에 불편을 느꼈다.  다행히 오른쪽에 미리보기를 통해 바로 인지를 할 수 있지만 텍스트를 치는 부분과 오른쪽의 미리보기 하는 부분이 일치하지 않을 수 있다.  

![이미지](https://ifh.cc/g/HwSNtH.png)  

이미지와 같은 상황을 말하는 것이다. 사용하는 내내 익숙하지 않은 것에 대해 불편을 느꼈다. 이에 따라 왜 마크다운을 해야하는 지에 대해 추가 조사를 했다.

마크다운은 일반 텍스트 기반의 경량 마크업 언어다. 마크업 언어는 양식이 있는 문서의 한 종류이며 그 양식을 태그 등으로 구분해 문서 정보를 정확하게 전달하는 문서이다. 마크업 언어는 여러 마크업 언어를 통틀어 말한다.  


#### 장점
---  
* 문법이 간결하고 쉬움  
* 대부분의 것에 사용 가능  
* 마크다운 지원하는 플랙폼 다양  
* 작성 및 수정하기 쉬움

* 형태 변환이 자유로움  
* 버전관리가 가능함
